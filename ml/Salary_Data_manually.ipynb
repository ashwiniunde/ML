```python
# Step 1Importing libraries 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
plt.scatter(x=df["YearsExperience"],y=df["Salary"])
plt.xlabel("YearaExperience")
plt.ylabel("Salary")
plt.title("Scatter plot of Years of Experience and Salary ")
plt.show()
```

```python
x_mean=df["YearsExperience"].mean()
x_mean
```

```python
y_mean=df["Salary"].mean() 
y_mean
```

```python
# Manually calculate slope (m)
numerator = 0
denominator = 0

for x, y in zip(df["YearsExperience"], df["Salary"]):
    numerator += (x - x_mean) * (y - y_mean)
    denominator += (x - x_mean) ** 2

m = numerator / denominator
```

```python
print("Slope (m) :",m)
```

```python
c=y_mean - m *x_mean
print("Intercept is : ",c)
```

```python

y_pred = m * df["YearsExperience"]+c
y_pred
```

```python
plt.plot(df["YearsExperience"],y_pred,color="blue")
plt.show()
```

```python
new_x = 2
```

```python
y_pred = m*new_x + c
```

```python
y_pred
```

```python
plt.scatter(x=df["YearsExperience"],y=df["Salary"])
plt.plot(df["YearsExperience"],y_pred,color="red")
plt.xlabel("YearaExperience")
plt.ylabel("Salary")
plt.title("Scatter plot of Years of Experience and Salary ")
plt.show()
```

```python

# Error Metrics
mse = np.mean((df["Salary"] - y_pred) ** 2)
rmse = np.sqrt(mse)
mae = np.mean(abs(df["Salary"] - y_pred))
r2 = 1 - (np.sum((df["Salary"] - y_pred) ** 2) / np.sum((df["Salary"] - y_mean) ** 2))
```

```python
print(f"MSE: {mse:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"MAE: {mae:.4f}")
print(f"R Score: {r2:.4f}")
```
